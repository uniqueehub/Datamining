```python
import pandas as pd

#loading the dataset in df variable
df=pd.read_csv("Heart.csv")
print(df)    #display the data in dataset
```

```python
# e]
print(df.dtypes)   # print datatypes in dataset
```

```python
# a]
#Pandas Series.isnull() function detect missing values in the given series object.
# detect missing values
df1=df.isnull()
print(df1)     #True means that particular value is null
print(df.isnull().sum())  #gives number of missing values in each attribute
totalNull=df.isnull().sum().sum()     #gives total missing values in dataset
print("TotalMissingValues = {}".format(totalNull))
```

```python
df2=df.fillna(value=5)   #fill all null values with 5
print(df2)
TM=df2.isnull().sum().sum()
print("TotalMissingValues = {}".format(TM))
```

```python
df3=df.fillna(method='pad') #pad values with last row
print(df3)
TM=df3.isnull().sum().sum()
print("TotalMissingValues = {}".format(TM))
```

```python
df4=df.bfill()   #replace null values with values from the next row
print(df4)
TM=df4.isnull().sum().sum()
print("TotalMissingValues = {}".format(TM)) # 1 missing value because the last Ca value is already empty, so it next row by default consider as empty
```

```python
df4=df.fillna(value=df['Ca'].mean())   #fill missing values with mean of Ca column
print(df4)
```

```python
df5=df.fillna(value=df['Ca'].std())  # fill missing values with standard deviation
print(df5)
TM1=df5.isnull().sum().sum()
print("TotalMissingValues = {}".format(TM1))
```

```python
df6=df.fillna(value=df['Ca'].min())   #fill missing values with minimun values in Ca column
print(df6)
```

```python
df7=df.fillna(value=df['Ca'].max())   #fill missing values with maximum value in Column
print(df7)
TM=df7.isnull().sum().sum()
print("TotalMissingValues = {}".format(TM))
```

```python
# a]
df8=df.dropna(how='all')  #when all values in record are null then tuple is droped
print(df8)
shape=df8.shape
print("Shape = {}".format(shape))
df9=df.dropna(how='any')  # when any one of the value is null then tuple is droped
print(df9)
shape=df9.shape
print("Shape = {}".format(shape))
df10=df.fillna(value=df['Age'].min())
print(df10)
df11=df.fillna(value=df['Age'].mean())
print(df11)
```

```python
df12=df.dropna()   #drop column having null value
print(df12)
shape=df12.shape
print("Shape = {}".format(shape))
```

```python
# c]
import pandas as pd
from scipy import stats

# Load the dataset
df = pd.read_csv('Heart.csv')

# Define a function to detect outliers using z-score
def detect_outliers_zscore(data):
    outliers = pd.DataFrame()
    for column in data.select_dtypes(include=['number']).columns:
        z_scores = stats.zscore(data[column])
        outliers[column] = data[abs(z_scores) > 3][column]
    return outliers

# Call the function to detect outliers
outliers = detect_outliers_zscore(df)

# Display the outliers found in each attribute
print("Outliers in each attribute:")
print(outliers)
```

```python
# c]
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Heart.csv')

# Select numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns

# Create boxplots for each numerical attribute
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    plt.title(f'Boxplot of {column}')
    plt.boxplot(df[column].dropna())  # Drop NA values for plotting
    plt.xlabel(column)
    plt.ylabel('Values')
    plt.show()
```

```python
# d]
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Heart.csv')

# Select the attributes for the histogram (e.g., age and cholesterol)
attribute1 = 'Age'
attribute2 = 'Chol'

# Plot histogram for attribute1 (Age)
plt.figure(figsize=(10, 6))
plt.hist(df[attribute1].dropna(), bins=20, color='lavender', edgecolor='black', alpha=0.7)
plt.title(f'Histogram of {attribute1}')
plt.xlabel(attribute1)
plt.ylabel('Frequency')
plt.show()

# Plot histogram for attribute2 (Cholesterol)
plt.figure(figsize=(10, 6))
plt.hist(df[attribute2].dropna(), bins=20, color='orange', edgecolor='black', alpha=0.7)
plt.title(f'Histogram of {attribute2}')
plt.xlabel(attribute2)
plt.ylabel('Frequency')
plt.show()
```
